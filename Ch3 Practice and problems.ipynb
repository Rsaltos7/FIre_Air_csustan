```python
#3.1
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the text file
data = pd.read_csv('sunspots.txt', delim_whitespace=True, header=None, names=['Month', 'Sunspots'])

# Check the first few rows of the DataFrame
print(data.head())

# Part a: Plot the sunspots as a function of time
plt.figure(figsize=(10, 5))
plt.plot(data['Month'], data['Sunspots'], label='Sunspots')
plt.xlabel('Month (Since Jan 1749)')
plt.ylabel('Number of Sunspots')
plt.title('Sunspots over Time')
plt.legend()
plt.show()

# Part b: Display only the first 1000 data points
plt.figure(figsize=(10, 5))
plt.plot(data['Month'][:1000], data['Sunspots'][:1000], label='Sunspots (First 1000 Points)', color='blue')
plt.xlabel('Month (Since Jan 1749)')
plt.ylabel('Number of Sunspots')
plt.title('Sunspots over Time (First 1000 Data Points)')
plt.legend()
plt.show()

# Part c: Calculate the running average
def running_average(data, r):
    return [sum(data[max(0, i-r):min(len(data), i+r+1)]) / (2*r + 1) for i in range(len(data))]

r = 5
sunspot_avg = running_average(data['Sunspots'][:1000], r)

# Plot original data and running average
plt.figure(figsize=(10, 5))
plt.plot(data['Month'][:1000], data['Sunspots'][:1000], label='Sunspots (First 1000 Points)', color='blue')
plt.plot(data['Month'][:1000], sunspot_avg, label='Running Average', color='orange')
plt.xlabel('Month (Since Jan 1749)')
plt.ylabel('Number of Sunspots')
plt.title('Sunspots and Running Average (First 1000 Data Points)')
plt.legend()
plt.show()
```

```python
#3.2
import numpy as np
import matplotlib.pyplot as plt

# Part a: Deltoid Curve
# Define theta values from 0 to 2π
theta_deltoid = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for the deltoid curve
x_deltoid = 2 * np.cos(theta_deltoid) + np.cos(2 * theta_deltoid)
y_deltoid = 2 * np.sin(theta_deltoid) - np.sin(2 * theta_deltoid)

# Plot the deltoid curve
plt.figure(figsize=(8, 8))
plt.plot(x_deltoid, y_deltoid, label='Deltoid Curve')
plt.title("Deltoid Curve")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()

# Part b: Galilean Spiral
# Define theta values from 0 to 10π
theta_spiral = np.linspace(0, 10 * np.pi, 1000)

# Calculate r and convert to Cartesian coordinates
r_spiral = theta_spiral**2
x_spiral = r_spiral * np.cos(theta_spiral)
y_spiral = r_spiral * np.sin(theta_spiral)

# Plot the Galilean spiral
plt.figure(figsize=(8, 8))
plt.plot(x_spiral, y_spiral, label='Galilean Spiral: r = θ²', color='orange')
plt.title("Galilean Spiral")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()

# Part c: Fey’s Function
# Define theta values from 0 to 24π
theta_fey = np.linspace(0, 24 * np.pi, 1000)

# Calculate r
r_fey = (np.exp(np.cos(theta_fey)) - 2 * np.cos(4 * theta_fey) + np.sin(theta_fey)**5) / 12

# Convert to Cartesian coordinates
x_fey = r_fey * np.cos(theta_fey)
y_fey = r_fey * np.sin(theta_fey)

# Plot Fey's function
plt.figure(figsize=(8, 8))
plt.plot(x_fey, y_fey, label="Fey's Function", color='green')
plt.title("Fey's Function")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

```python
#3.5
import matplotlib.pyplot as plt

def plot_nacl_lattice(size):
    plt.figure(figsize=(8, 8))
    for i in range(size):
        for j in range(size):
            # Sodium atoms (blue)
            plt.scatter(i, j, color='blue', s=100)  # Na
            # Chlorine atoms (green)
            plt.scatter(i + 0.5, j + 0.5, color='green', s=100)  # Cl
    plt.title('Sodium Chloride Lattice')
    plt.xlim(0, size)
    plt.ylim(0, size)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid()
    plt.show()
plot_nacl_lattice(10)
def plot_fcc_lattice(size):
    plt.figure(figsize=(8, 8))
    for i in range(size):
        for j in range(size):
            for k in range(size):
                # Corner atoms (1)
                plt.scatter(i, j, color='orange', s=100)
                # Face-centered atoms (0.5, 0.5, k) positions
                if i == 0.5 or j == 0.5:
                    plt.scatter(i + 0.5, j + 0.5, color='orange', s=100)
    plt.title('Face-Centered Cubic Lattice')
    plt.xlim(0, size)
    plt.ylim(0, size)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid()
    plt.show()
plot_fcc_lattice(10)
```

```python
#3.6
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r_values = np.arange(1, 4, 0.01)  # Values of r from 1 to 4
num_iterations = 2000  # Total iterations for each r
last_iterations = 1000  # Last iterations to plot (to allow settling)

# Array to hold the final values of x for each r
x_final = np.zeros(len(r_values))

# Iterate over each value of r
for i, r in enumerate(r_values):
    x = 0.5  # Starting value of x
    # Perform initial iterations to allow x to settle
    for _ in range(num_iterations):
        x = r * x * (1 - x)
    
    # Now, run the last iterations and store the values to plot
    for _ in range(last_iterations):
        x = r * x * (1 - x)
        x_final[i] = x  # Store the last value of x for this r

# Create the Feigenbaum plot
plt.figure(figsize=(12, 6))
plt.plot(r_values, x_final, 'k.', markersize=1)  # Plot with dots
plt.title('Feigenbaum Plot')
plt.xlabel('r')
plt.ylabel('x')
plt.xlim(1, 4)
plt.ylim(0, 1)
plt.grid()
plt.show()
```

```python
#3.7
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the Mandelbrot set
def mandelbrot(N, max_iter):
    # Create a grid of complex numbers
    x = np.linspace(-2, 2, N)
    y = np.linspace(-2, 2, N)
    C = np.array([[complex(i, j) for i in x] for j in y])  # Create complex grid
    Z = np.zeros(C.shape, dtype=complex)  # Initialize Z array
    output = np.zeros(C.shape, dtype=int)  # Array to hold iteration counts

    # Iterate over the grid
    for i in range(max_iter):
        mask = np.abs(Z) <= 2  # Mask for values that haven't escaped
        Z[mask] = Z[mask] ** 2 + C[mask]  # Apply the Mandelbrot iteration
        output[mask] = i  # Record the iteration count for points still within the circle

    return output

# Parameters
N = 1000  # Resolution of the grid
max_iter = 100  # Maximum number of iterations

# Calculate the Mandelbrot set
output = mandelbrot(N, max_iter)

# Create a density plot
plt.figure(figsize=(10, 10))
plt.imshow(output, extent=(-2, 2, -2, 2), cmap='hot', interpolation='bilinear')
plt.colorbar(label='Number of Iterations')
plt.title('Mandelbrot Set')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```

```python
#3.8
import numpy as np
import matplotlib.pyplot as plt

# Part a: Read data from file
data = np.loadtxt("millikan.txt", float)
x = data[:, 0]  # Frequencies (ν) in Hz
y = data[:, 1]  # Voltages (V) in Volts

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Part b: Calculate Ex, Ey, Exx, Exy and find slope (m) and intercept (c)
N = len(x)
Ex = np.sum(x) / N
Ey = np.sum(y) / N
Exx = np.sum(x**2) / N
Exy = np.sum(x * y) / N

# Calculate slope (m) and intercept (c)
m = (Exy - Ex * Ey) / (Exx - Ex**2)
c = (Exx * Ey - Ex * Exy) / (Exx - Ex**2)

print(f"Slope (m): {m:.6e}")
print(f"Intercept (c): {c:.6e}")

# Part c: Evaluate m*x + c for each data point and plot the best fit line
x_fit = np.linspace(min(x), max(x), 100)  # Generate x values for the fit line
y_fit = m * x_fit + c  # Calculate corresponding y values

plt.plot(x_fit, y_fit, color='red', label='Best Fit Line')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Voltage (V)")
plt.title("Photoelectric Effect Data and Best Fit Line")
plt.legend()
plt.grid()
plt.show()

# Part d: Calculate Planck's constant using the slope
# Using the equation V = (h/e) * ν - φ
# Rearranging gives: h/e = m, so h = m * e
e = 1.602e-19  # Charge of the electron in Coulombs
h = m * e  # Calculate Planck's constant
print(f"Calculated Planck's constant (h): {h:.6e} J·s")

# Accepted value of Planck's constant
h_accepted = 6.626e-34  # J·s
percent_difference = abs((h - h_accepted) / h_accepted) * 100
print(f"Percent difference from accepted value: {percent_difference:.2f}%")
```
